```python
# imports here
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

import time
import os
from config import USERNAME,PWD,PAGE_URL,SCROLL_N
```

```python

if not (USERNAME and PWD and PAGE_URL):
    raise Exception('Username , password or page url not set')


chrome_options = webdriver.ChromeOptions()
prefs = {"profile.default_content_setting_values.notifications" : 2}
chrome_options.add_experimental_option("prefs",prefs)
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=chrome_options)
```

```python
# LOGIN
driver.get("http://www.facebook.com")

username = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='email']")))
password = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='pass']")))

username.clear()
username.send_keys(USERNAME)
password.clear()
password.send_keys(PWD)

button = WebDriverWait(driver, 2).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button[type='submit']"))).click()
```

```python
# open and scroll down
driver.get(f"{PAGE_URL}photos/")
time.sleep(5)
for j in range(0, SCROLL_N):
    print("scrolling")
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)
```

```python
anchors = driver.find_elements(By.TAG_NAME, "a")
anchors = [a.get_attribute("href") for a in anchors]
anchors = [a for a in anchors if str(a).startswith(f"{PAGE_URL}photos")]
print("Found " + str(len(anchors)) + " links to images")
```

```python
images = [] 



# extract the [1]st image element in each link
images_found = 0
for a in anchors:
    images_found += 1
    driver.get(a) #navigate to link
    time.sleep(5) #wait a bit
    img = driver.find_elements(By.TAG_NAME, "img")

    current_img = img[0].get_attribute("src")
    images.append(current_img)  # may change in future to img[?]

    print(f"images_found {images_found}")


print('I scraped '+ str(len(images)) + ' images!')
```

### Step 5: Save all the photos in the new directory

Using wget

```python
import wget

#download images
counter = 0
for image in images:
    save_as = os.path.join(path, str(counter) + '.jpg')
    wget.download(image, save_as)
    counter += 1
```
